###     __getattribute__ -   ,          .

```python
class Person:
    def __getattr__(self, item):
        return f" {item}  "

person = Person()
print(person.age)  # :  age  
```

```python
class Person:
    def __getattr__(self, name):
        if name == "age":
            return 30
        else:
            return "  "

person = Person()
print(person.age)   # : 30
print(person.name)  # :   
```

```python
class Person:
    def __init__(self):
        self.attributes = {"name": "John", "age": 30}

    def __getattr__(self, name):
        if name in self.attributes:
            return self.attributes[name]
        else:
            return "  "

person = Person()
print(person.name)   # : John
print(person.age)    # : 30
print(person.height) # :   
```

```python
class Person:
    def __getattr__(self, item):
        if item == 'name':
            self.name = 'Vasya'
            return self.name
        else:
            return '  '
        
        
person = Person()
print(person.name)
```
